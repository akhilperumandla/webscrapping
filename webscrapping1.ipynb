```python
import pandas as pd
import requests
import csv
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=laptop&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
response=requests.get(url)
```

```python
response
```

```python
link='ww.flipkart.com/robots.txt'
```

```python
link
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
product=[]
price=[]
rating=[]
product_data=soup.findAll('div',attrs={'class':'_4rR01T'})
print(product_data)
```

```python
product_data
```

```python
for a in soup.findAll('a',href=True,attrs={'class':'_1fQZEK'}):
    laptop_name=a.find('div',attrs={'class':'_4rR01T'}).text.replace(')','')
    prices=a.find('div',attrs={'class':'_30jeq3_1_WHN1'})
    ratings=a.find('div',attrs={'class':'_3LWZ1K'})
    product.append(laptop_name)
    price.append(prices)
    rating.append(ratings)
```

```python
info=(product,price,rating)
print(info)
```

```python
laptop_dataset=pd.DataFrame({'product':product,'prices':prices,'ratings':ratings})
```
